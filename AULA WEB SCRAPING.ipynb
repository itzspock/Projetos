```python
import requests
from bs4 import BeautifulSoup
```

```python
mercado_livre = "https://lista.mercadolivre.com.br/gabinete-full-tower-atx#D[A:gabinete%20full%20tower%20atx,L:undefined]"
```

```python
mercado_livre
```

```python
resposta = requests.get (mercado_livre)
```

```python
resposta
```

```python
resposta.text
```

```python
#Parser -> Analisador de TEXTO

site = BeautifulSoup (resposta.text, "html.parser")
```

```python
site
```

```python
info = site.find_all ("div", attrs = {"class" : "ui-search-result__content-wrapper"})
```

```python
info
```

```python
for X in info:
    print(X)
```

```python
for item in info:
    print(item.find_all("h2"))
```

```python
for item in info:
    print(item.find_all("h2")[0].text)
```

```python
produto = []
valor = []

for item in info:
    nome=item.find_all("h2")[0].text
    produto.append (nome)
    print(nome)

    preco = (item.find_all ("span", attrs = {"class": "andes-money-amount__fraction"})[0].text)
    valor.append (preco)
    print(preco)
```

```python
base= {
    "item" : produto,
    "preco" : valor
}
```

```python
base
```
